```python
import marvin
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the data cube of flux
cube_data= marvin.tools.Cube(filename='manga-8319-9101-LOGCUBE.fits.gz')
flux =  cube_data.flux  # np.transpose(cube_data.flux, axes=(2,1,0))
```

```python
#Select a slice of flux containing only few 100s of wavelgnths
flux_slice=flux[240:2600, :, :]
```

```python
#Do this step to convert marvin array to numpy array
summed_flux_slice=flux_slice.sum(axis=0).value
```

```python
#Plot the selected slice
plt.imshow(summed_flux_slice)
```
