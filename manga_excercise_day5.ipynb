```python
import marvin
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data cube of flux and get maps for different quantities
cube_data= marvin.tools.Cube(filename='manga-7968-12702-LOGCUBE.fits.gz')
map_cube=cube_data.getMaps()
```

```python
#Change the shape of the different items of the data cube to from 0,1,2 to 2,1,0
flux =  cube_data.flux
```

```python
#Select a slice of flux containing only few 100s of wavelgnths
flux_slice=flux[3000:3600, :, :]

#Do this step to convert marvin array to numpy array
summed_flux_slice=flux_slice.sum(axis=0).value

#Plot the selected slice
plt.imshow(summed_flux_slice, vmax=500, vmin=0, cmap='twilight') #viridis, coolwarm, plasma, cividis, inferno
```

```python
#Fetch map for Halpha line
ha_line_map= map_cube.emline_gflux_ha_6564.value

```
