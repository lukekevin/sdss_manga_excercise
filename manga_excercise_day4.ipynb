```python
import marvin
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data cube of flux and get maps for different quantities
cube_data= marvin.tools.Cube(filename='manga-8319-9101-LOGCUBE.fits.gz')
map_cube=cube_data.getMaps()
```

```python
#Map of O3 line
o3_line_map=map_cube.emline_gflux_oiii_5008.value
```

```python
plt.imshow(o3_line_map, vmax=0.5, vmin=0)
```

```python
#Map of stellar sigma
stellar_sigma=map_cube['stellar_sigma'].value
```

```python
plt.imshow(stellar_sigma, vmax=100, vmin=0)
```

```python
#Map of Hbeta line
hbeta_map=map_cube.getMap('emline_gflux', 'H-beta').value
```

```python
plt.imshow(hbeta_map, vmax=0.5, vmin=0)
```
